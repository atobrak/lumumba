# importing the spectrogram dataset



```python
import numpy as np
import matplotlib.pyplot as plt

import os
from os import walk
import matplotlib.image as mpimg
import pandas as pd
```

```python
# importing the otherimportant features



col_list = ["timestamp","Epicenter Elevation [km]", "Hypocentral Elevation [km]","Travel Distance [km]","Magnitude","Depth [km]"]
af1 = pd.read_csv("oklahoma_induced_1.csv",  sep =',',usecols=col_list)
af2 = pd.read_csv("california_tectonic_1.csv",  sep =',',usecols=col_list)
print(af1)
print(af2)
```

# Induced Earthquakes

```python
# changing the working directory 

pname = 'c:\spectrograms\Oklahoma\Channel 1'
```

```python
# reading the list of waveforms from the working directory

for dirpath,dirnames, filenames in walk (pname):
    print(filenames)
```

```python
####### reading out files from your directory

l_img = [None] * len(filenames)
shape0 = None
##### iterating over the given list
for i, f in enumerate(filenames):
    fp = os.path.join(dirpath,f)
    ### reading the files
    img = mpimg.imread(fp)
    l_img[i] = img.copy()
    if i == 0:
        shape0 = img.shape
    else:
        if shape0 != img.shape:
            print('Not all images are of the same shape. First: %s. i-th: %s' % (shape0, img.shape))
    if i < 2:
        print('min, max: [%.1f, %.1f]' % (np.min(img), np.max(img)))
        #print(img)
        plt.imshow(img)
        plt.show()
num_samples = len(l_img)
```

### You will want to try and extract raw info from the spectograms. 

```python
bI=af1['timestamp']
cI=bI.to_numpy()
bT=af2['timestamp']
cT=bT.to_numpy()
```

```python
# Next, we need an elaborate dataframe containing all relevant information for each sample we have. This will require more thna just the city name + filename

# feat_names = ['city', 'channel_id', 'label_is_induced', 'spectogram', 'timestamp', 'etc.']
feat_dict = dict()
feat_dict['city'] = np.tile('Oklahoma', reps=(num_samples))
feat_dict['channel_id'] = np.tile(1, reps=(num_samples))





feat_dict['timestamp'] = cI # This needs to be actual timestamps and not just random integers.

# what is the best way to input the actual time stamp here. the goal is to have the same timestamps here and in the csv file above
# so that i can merge the two data frames. the csv contains the other important features such as magnitude, depth etc.


feat_dict['label_is_induced'] = np.tile(True, reps=(num_samples))
feat_dict['spectogram_filepath'] = [os.path.join(dirpath,f) for f in filenames]
feat_dict['spectogram'] = [lambda: l_img[i] for i in range(num_samples)]




# feat_dict['...'] # ... Continue. Examples include the statistics you had in mind, some other details on the spectograms, etc. 

df_I1 = pd.DataFrame(data=feat_dict, index=feat_dict['timestamp'])
display(df_I1)
```

```python
df_merge_col_I1 = pd.merge(df_I1, af1, on='timestamp')

df_merge_col_I1
```
